```python
import pandas as pd
from datetime import datetime, date
```

```python
bikes_2019 = pd.read_csv('201909-citibike-tripdata.csv')
bikes_2019.head()
```

```python
bikes_2020 = pd.read_csv('JC-202009-citibike-tripdata.csv')
bikes_2020.head()
```

```python
#Change birth year to age for 2020 data
current_year = 2023
age_2020 = bikes_2020['birth year']
age = current_year - age_2020
bikes_2020['age'] = age
```

```python
#Change birth year to age for 2019 data
current_year = 2023
age_2019 = bikes_2019['birth year']
age = current_year - age_2019
bikes_2019['age'] = age
```

```python
#Change number to gender for 2019 data
bikes_2019.loc[bikes_2019["gender"] == 1, "gender"] = "Male"
bikes_2019.loc[bikes_2019["gender"] == 2, "gender"] = "Female"
bikes_2019.loc[bikes_2019["gender"] == 0, "gender"] = "Unknown"
bikes_2019.head()
```

```python
#Change number to gender for 2020 data
bikes_2020.loc[bikes_2020["gender"] == 1, "gender"] = "Male"
bikes_2020.loc[bikes_2020["gender"] == 2, "gender"] = "Female"
bikes_2020.loc[bikes_2020["gender"] == 0, "gender"] = "Unknown"
bikes_2020.head()
```

```python
# Outputting both files to csv in Outputs folder
bikes_2019.to_csv("bikes_2019_clean.csv", index=False, header=True)
```
